```python
#Creating US 2019 pre-COVID table with relevant insight columns
us_19 = pd.read_csv(allcsv[-3])
us_19.head()

pre_covid_us = us_19[['Which of the below titles best describes your core discipline?', 'Throughout your career to date, what is the average length of time you have stayed in a role?']]
column_name_5 = {'Which of the below titles best describes your core discipline?':'Core Discipline',
               'Throughout your career to date, what is the average length of time you have stayed in a role?':'US 2019 - Avg length of time in role'}
pre_covid_us = pre_covid_us.rename(columns=column_name_5)
pre_covid_us

#Drop NaNs
pre_covid_us['US 2019 - Avg length of time in role'].isnull().sum(axis = 0)
pre_covid_us = pre_covid_us.dropna(subset=['US 2019 - Avg length of time in role'])
pre_covid_us['US 2019 - Avg length of time in role'].isnull().sum(axis = 0)

#Calculate means
pre_covid_us = pre_covid_us.groupby(['Core Discipline']).mean()
pre_covid_us = pre_covid_us.sort_values('US 2019 - Avg length of time in role', ascending=False)
pre_covid_us.head()
```

```python
us_21 = pd.read_csv(allcsv[-1])
us_21.head()

post_covid_us = us_21[['Which of the below titles best describes your core discipline?', 'Throughout your career to date, what is the average length of time you have stayed in a role?']]
column_name_6 = {'Which of the below titles best describes your core discipline?':'Core Discipline',
               'Throughout your career to date, what is the average length of time you have stayed in a role?':'US 2021 - Avg length of time in role'}
post_covid_us = post_covid_us.rename(columns=column_name_6)
post_covid_us

#Drop NaNs
post_covid_us['US 2021 - Avg length of time in role'].isnull().sum(axis = 0)
post_covid_us = post_covid_us.dropna(subset=['US 2021 - Avg length of time in role'])
post_covid_us['US 2021 - Avg length of time in role'].isnull().sum(axis = 0)

#Calculate means
post_covid_us = post_covid_us.groupby(['Core Discipline']).mean()
post_covid_us = post_covid_us.sort_values('US 2021 - Avg length of time in role', ascending=False)
post_covid_us.head()
```

```python
#Combining pre-Covid Tables
area = ['UK', 'EU', 'US']
pre_covid_avg = []
post_covid_avg = []

pre_covid_avg.append(pre_covid_uk['UK 2019 - Avg length of time in role'].mean())
pre_covid_avg.append(pre_covid_eu['EU 2019 - Avg length of time in role'].mean())
pre_covid_avg.append(pre_covid_us['US 2019 - Avg length of time in role'].mean())

post_covid_avg.append(post_covid_uk['UK 2021 - Avg length of time in role'].mean())
post_covid_avg.append(post_covid_eu['EU 2021 - Avg length of time in role'].mean())
post_covid_avg.append(post_covid_us['US 2021 - Avg length of time in role'].mean())
post_covid_avg
pre_covid_avg

role_time = pd.DataFrame()
role_time['Area'] = area
role_time['Pre-COVID average years in role'] = pre_covid_avg
role_time['Post-COVID average years in role'] = post_covid_avg
role_time
```

```python
#Data visualistation
import seaborn as sns
import matplotlib.pyplot as plt

role_time.groupby(['Area']).mean().plot(kind ='barh')
plt.title("Average length of years in role across UK, EU and US (Pre-Covid & Post-Covid)")
plt.xlabel('Average Years in Role')
```

```python
#Extracting relevant data for UK dataset
female_uk = uk_21[["Which of the below titles best describes your core discipline?", "What percentage of your company's leadership team is female?"]]
column_1 = {'Which of the below titles best describes your core discipline?':'Core Discipline',
            "What percentage of your company's leadership team is female?":'UK Female Percentage of Leadership Team'}
female_uk = female_uk.rename(columns=column_1)

#female_uk.info()
#Drop irrelavant data
female_uk['UK Female Percentage of Leadership Team'].isnull().sum(axis = 0)

female_uk.drop(female_uk[female_uk['UK Female Percentage of Leadership Team'] == "not applicable"].index, inplace=True)
female_uk['UK Female Percentage of Leadership Team'].max()
female_uk['UK Female Percentage of Leadership Team'].min()

#Change data type into float
female_uk['UK Female Percentage of Leadership Team'] = female_uk['UK Female Percentage of Leadership Team'].astype(float)

#Mean calculations
female_uk = female_uk.groupby(['Core Discipline']).mean()
female_uk = female_uk.sort_values('UK Female Percentage of Leadership Team', ascending=False)
female_uk.head()

```
